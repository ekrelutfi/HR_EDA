```python
df_active = df[df['EmploymentStatus']=='Active']
```

```python
plt.figure(figsize=(10,7))
plt.tight_layout()
ax = sns.countplot(df_active['EmpStatusID'])
plt.title('Employee_status_ID', fontsize=22)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.legend(loc='upper right')
#plt.savefig('emp_status_id.png',bbox_inches="tight", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.tight_layout()
ax = sns.countplot(df_active['EmpStatusID'], hue=df_active['RoleID'])
plt.title('Employee_status_ID', fontsize=22)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.legend(loc='upper right')
#plt.savefig('emp_status_id.png',bbox_inches="tight", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.tight_layout()
ax = sns.countplot(df_active['Tribe'])
plt.title('Tribe', fontsize=22)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
#plt.legend(loc='upper right')
#plt.savefig('Tribe.png',bbox_inches="tight", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.tight_layout()
ax = sns.countplot(y=df_active['Squad'])
plt.title('Squad', fontsize=22)

total = len(df_active['Squad'])
for p in ax.patches:
        percentage = '{:.0f}'.format(p.get_width())
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
        
#plt.savefig('Squad.png',bbox_inches="tight", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,16))
sns.heatmap(df_active.groupby(['Squad','RoleID'])['EmpID'].count().unstack().reset_index().fillna(0).set_index('Squad'), cmap='Blues', annot=True)
plt.tight_layout()
#plt.savefig('squad role.png', dpi=300, )
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(df_active.groupby(['RoleID','Squad'])['EmpID'].count().unstack().reset_index().fillna(0).set_index('RoleID'), cmap='Blues', annot=True)
plt.title('Squad, RoleID')
plt.tight_layout()
#plt.savefig('role_squad.png', dpi=300, )
plt.show()
```

```python
plt.figure(figsize=(16,9))
g=sns.heatmap(df_active.groupby(['Squad','Tribe'])['EmpID'].count().unstack().reset_index().fillna(0).set_index('Squad'), cmap='Blues')
g.set_yticklabels(g.get_yticklabels(), rotation = 0, fontsize = 8)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df_active.groupby(['RoleID','Tribe'])['EmpID'].count().unstack().reset_index().fillna(0).set_index('RoleID'), cmap='Blues', annot=True)
plt.title('RoleID,Tribe')
plt.tight_layout()
#plt.savefig('role_tribe.png', dpi=300, )

plt.show()
```

```python
plt.figure(figsize=(10,7))
ax=sns.countplot(df_active['RoleID'])
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.title('RoleID')
plt.xticks(rotation=90)
plt.tight_layout()
#plt.savefig('role.png', dpi=300, )

plt.show()
```
